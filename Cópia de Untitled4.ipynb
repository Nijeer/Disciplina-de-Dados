```python
#importando as bbibliotecas
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/supermarket_sales (1).csv', parse_dates=['Date'])
#criando um novo df agrupado por cidades
df_cidades = df.groupby('City').sum()
df_cidades
```

```python
#dados de eixo x 
x = df_cidades.index
#dados do eixo y
y = df_cidades['Total']
```

```python
from matplotlib.projections import ProjectionRegistry
fig, ax = plt.subplots(figsize=(10, 6))
barras = ax.bar(x, y, 0.5, label='Vendas')
ax.set_title("Vendas por cidades", fontname='Arial', fontsize=18, color='r')
ax.set_ylabel("Valores em R$")
ax.set_xlabel("Cidades");
print(barras[0].get_height())
print(barras[1].get_height())
print(barras[2].get_height())
for barra in barras:
  rotulo = f"{barra.get_height():_.2f}"
  rotulo = rotulo.replace('.',',')
  rotulo = rotulo.replace('_','.')
  posx = barra.get_x()
  posy = barra.get_height()
  larg = barra.get_width()
  ax.annotate(rotulo, xy=(posx+larg/2,posy), ha='center', xytext=(0,5), textcoords=('offset points'))
```
