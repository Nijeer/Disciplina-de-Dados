```python
#criar um dicionario vazio
avengers = {}#criar um dicionario vazio
print(avengers)#apresentar o dicionario
```

```python
#adicionar itens ao dicionario
avengers['vingador1']='capitão america'
avengers['vingador2']='campiã marvel'
avengers['vingador3']='homem de ferro'
avengers['vingador4']='hulk'
avengers['vingador5']='thor'
print(avengers)
```

```python
#acessar um item no dicionario
print(avengers['vingador1']) 
print(avengers.get('vingador2'))
print(avengers.get('vimgador10'))
```

```python
#alterar o valor de um item do dicionario
avengers['vingador3']='viuva negra'
print(avengers)
```

```python
#excluir um item do dicionario
del avengers['vingador3']
print(avengers)
```

```python
#outros metodos de tipo de dicionario
print(len(avengers))
print(avengers.keys())
print(avengers.items())
print(avengers.values())
avengers.pop('vingador1')
avengers.clear()
print(avengers)
```
