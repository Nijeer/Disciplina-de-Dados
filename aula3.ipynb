```python
#Criando uma lista com intervalos numéricos 
numeros = list(range(10))
print(numeros)
numeros = list(range(1,13,2))
print(numeros)
numeros_impares = list(range(1,100,2))
print(numeros_impares)
mumeros_pares = list(range(2,100,2))
print(mumeros_pares)
```

```python
#acessar elementos pelo indici
elementos = ['POO','BD','logica' ,'redes', 'APS', 'fundamentos']
print(elementos)
print(len(elementos))
print(elementos[1])
print(elementos[1:3])
print(elementos[1:5:2])

```

```python
#criar uma nova lista
lcdp = ['Tokyo', 'Nairobi', 'Lisboa', 'Berlin', 'Rio', 'Denver','Helsink']
print(lcdp)
```

```python
#adicionar um elemento ao final da lista
lcdp.append('Marsells')
print(lcdp)
```

```python
#Adicionar um  elemento na posiçao escolhida
lcdp.insert(2, 'Moscow')
print(lcdp)
```

```python
#Ordenar uma lista
lcdp.sort()
print(lcdp)
```

```python
#ordenar uma lisa de forma inversa
lcdp.reverse()
print(lcdp)
```

```python
#Remover o ultimo elemento da lista
lcdp.pop()
print(lcdp)
```

```python
#remove um elemento pelo indici
lcdp.pop(1)
print(lcdp)
```

```python
#verificar se um elemento pertence a lista 
print('Tokyo' in lcdp)
print('Berlin' in lcdp)
```
