```python
peso = float(input("Digite seu peso: "))
altura = float(input("Digite sua altura: "))
imc = peso/altura**2
print("seu imc é: ",imc)
if imc >= 30:
  print("Obesidade")
elif imc >= 25:
  print("Sobrepso")
elif imc >= 18.5:
  print("Normal")
elif imc <18.5:
  print("Desnutrido")
```
