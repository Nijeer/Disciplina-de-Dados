```python
#criando um loop infinito
while True:
  nome=input("Digite o nome do aluno: ")
  print(nome)
```

```python
#interromper um loop
while True:
  nome=input("Digite o nome do aluno: ")
  print(nome)
  if nome == 'Sair':
    break
    
```

```python
#retornando para o inicio do loop
while True:
  nome=input("Digite o nome do aluno: ")
  if nome == 'Sair' :
    break
  elif nome == 'Alex':
      continue
      print(nome)
```

```python
#criando um loop com interaçao do usuario
resposta = 'S'
while resposta[0].upper() == 'S':
  numero=int(input("Digite um numero: "))
  print("O quadrado do numero",numero,"é",numero**2)
  resposta=input("Deseja informar um novo numero: ")
```

```python
#utilizando for com resposta do usuario
quantidade = int(input("Quantos alunos deseja cadastrar: "))
for x in range(quantidade):
  nome=input("Digite o nome de Aluno: ")
```

```python
#acumulando valores
soma = 0
for x in range(1,11):
  soma = soma + x
print(soma)
  
```

```python
#percorrendo diversas listas ao mesmo tempo
nomes = ["Carina","Tamar","Grace","Stefan","Walter","Lamarana","Alma","Gustavo","Raul","Valentina","Isaura","Iago","Malika","Martinha","Riana","Amora","Cauã","Luciano","Francis","José"]
apelidos = ["Jamandas","Amado","Filgueiras","Estrela","Alcaide","Fiúza","Furtado","Amaral","Guerreiro","Varela","Belmonte","Mainha","Sandinha","Loureiro","Sanches","Camargo","Noleto","Borges","Canto","Vilaverde"]
for i, (nome, apelido) in enumerate(zip(nomes, apelidos)):
  print(i+1,"Nome: ",nome,"Apelido: ",apelido)
```
