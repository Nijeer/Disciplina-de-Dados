```python
arq = open('/content/drive/MyDrive/Colab Notebooks/ocorrencia.csv')
dados = arq.read()
dados = dados.split('\n')
tipos = []
locais = []
datas = []
estados = []
for dado in dados:
  if dado != '':
   dado = dado.split(';')
   tipos.append(dado[5])
   locais.append(dado[8])
   datas.append(dado[12])
   estados.append(dado[9])

```

```python
for tipo, local, data, estado in zip(tipos, locais, datas, estados):
  print(tipo, local, data, estado)
```

```python
novo_arq = open('/content/drive/MyDrive/Colab Notebooks/ocorrencia2.csv','w')
for tipo, local, data in zip(tipos, locais, datas, estados):
  novo_arq.write(tipo+';'+local+';'+data+'\n')
novo_arq.close()
```
